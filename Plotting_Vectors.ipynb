```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.quiver(0,0,3,4)
plt.show()
```

```python
plt.quiver(0,0,3,4, scale_units='xy', angles='xy', scale=1)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.show()
```

```python
plt.quiver(0,0,3,4, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0,0,-3,4, scale_units='xy', angles='xy', scale=1, color='g')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid()
plt.show()
```

```python
def plot_vectors(vecs):
  color = ['r', 'g', 'b', 'y', 'p']
  i = 0
  for vec in vecs:
    plt.quiver(0,0,vec[0],vec[1], scale_units='xy', angles='xy', scale=1, color=color[i%len(color)])
    i += 1
  plt.xlim(-10, 10)
  plt.ylim(-10, 10)
  plt.grid()
  plt.show()
```

```python
vecs = [(1, 2), (5, 10), (-5, 3)]
plot_vectors(vecs)
```
