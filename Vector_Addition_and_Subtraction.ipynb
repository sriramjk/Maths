```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_vectors(vecs):
  color = ['r', 'g', 'b', 'y', 'p']
  i = 0
  for vec in vecs:
    plt.quiver(0,0,vec[0],vec[1], scale_units='xy', angles='xy', scale=1, color=color[i%len(color)])
    i += 1
  plt.xlim(-10, 10)
  plt.ylim(-10, 10)
  plt.grid()
  plt.show()
```

```python
a_vec = (1, 5)
b_vec = (-3, -5)
res = np.asarray(a_vec) + np.asarray(b_vec)
print(res)
plot_vectors([a_vec, b_vec, res])
```

```python
a_vec = (1, 5)
b_vec = (-3, -2)
res = np.asarray(a_vec) - np.asarray(b_vec)
print(res)
plot_vectors([np.asarray(a_vec), -1*np.asarray(b_vec), res])
```
